```python
def convert_cat(list, df):
    for feature in list:
        df[feature] = df[feature].astype("category")
category_fields = ["MSSubClass", "MSZoning", "Street", "Alley", "LandContour", "LotConfig",
                "Neighborhood", "Condition1", "Condition2", "BldgType", "HouseStyle", "RoofStyle",
                "RoofMatl", "Exterior1st", "Exterior2nd", "MasVnrType", "Foundation", "Heating",
                "CentralAir", "GarageType", "MiscFeature", "SaleType", "SaleCondition",
                'OverallQual', 'OverallCond', 'ExterQual', 'ExterCond', 'BsmtQual', 'BsmtCond',
                'HeatingQC', 'KitchenQual', 'FireplaceQu', 'GarageQual', 'GarageCond', 'PoolQC',
                'LotShape', 'LandSlope', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2', 'Functional',
                'GarageFinish', 'PavedDrive', 'Utilities', 'CentralAir', 'Electrical', 'Fence']

convert_cat(category_fields, train)
convert_cat(category_fields, test)
train.info()
```

```python
for name in train.select_dtypes("category"):
    # Check if "None" is already a category
    if "None" not in train[name].cat.categories:
        # Add "None" as a new category
        train[name] = train[name].cat.add_categories("None")
        test[name] = test[name].cat.add_categories("None")
    # Fill missing values with "None"
    train[name] = train[name].fillna("None")
    test[name] = test[name].fillna("None")
test.info()
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder


# Define transformers for numerical and categorical columns
numerical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])
```

```python
# Assuming you have already defined 'numerical_transformer' and 'categorical_transformer'

categorical_columns = train.select_dtypes(include=['object', 'category']).columns
numerical_columns = train.select_dtypes(include=['int64', 'float64']).columns
numerical_columns = numerical_columns.drop('SalePrice')

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_columns),
        ('cat', categorical_transformer, categorical_columns)
    ], remainder='passthrough')

# Create a pipeline with the preprocessor
pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor)])

# Apply the pipeline to your training dataset
X = train.drop('SalePrice', axis=1)
y = np.log(train['SalePrice'])  # normalize dependent variable
X_preprocessed = pipeline.fit_transform(X)

# Now, apply the same pipeline to your test dataset
X_test = test  # Assuming 'test' is your test dataset
X_test_preprocessed = pipeline.transform(X_test)
```

```python
y
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import Ridge
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.ensemble import StackingRegressor
from xgboost import XGBRegressor
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler

# Define the models
ridge_model = Ridge()
svr_linear_model = SVR(kernel='linear')
svr_rbf_model = SVR(kernel='rbf')
rf_model = RandomForestRegressor()
xgb_model = XGBRegressor(learning_rate=0.1, max_depth=3, n_estimators=500)

# Ensemble model using Stacking
stacked_model = StackingRegressor(
    estimators=[
        ('ridge', ridge_model),
        ('svr_linear', svr_linear_model),
        ('svr_rbf', svr_rbf_model),
        ('rf', rf_model),
        ('xgb', xgb_model)  # Add XGBoost to the ensemble
    ],
    final_estimator=GradientBoostingRegressor()
)

# Perform cross-validation
ridge_scores = cross_val_score(ridge_model, X_preprocessed, y, scoring='neg_mean_squared_error', cv=3, n_jobs=-1)
svr_linear_scores = cross_val_score(svr_linear_model, X_preprocessed, y, scoring='neg_mean_squared_error', cv=3, n_jobs=-1)
svr_rbf_scores = cross_val_score(svr_rbf_model, X_preprocessed, y, scoring='neg_mean_squared_error', cv=3, n_jobs=-1)
rf_scores = cross_val_score(rf_model, X_preprocessed, y, scoring='neg_mean_squared_error', cv=3, n_jobs=-1)
xgb_scores = cross_val_score(xgb_model, X_preprocessed, y, scoring='neg_mean_squared_error', cv=3, n_jobs=-1)
stacked_scores = cross_val_score(stacked_model, X_preprocessed, y, scoring='neg_mean_squared_error', cv=3, n_jobs=-1)

# Convert scores to root mean squared error (RMSE)
ridge_rmse = np.sqrt(-ridge_scores.mean())
svr_linear_rmse = np.sqrt(-svr_linear_scores.mean())
svr_rbf_rmse = np.sqrt(-svr_rbf_scores.mean())
rf_rmse = np.sqrt(-rf_scores.mean())
xgb_rmse = np.sqrt(-xgb_scores.mean())
stacked_rmse = np.sqrt(-stacked_scores.mean())

# Print the cross-validated RMSE
print(f'Ridge Regression Cross-Validated RMSE: {ridge_rmse}')
print(f'SVR Linear Kernel Cross-Validated RMSE: {svr_linear_rmse}')
print(f'SVR RBF Kernel Cross-Validated RMSE: {svr_rbf_rmse}')
print(f'Random Forest Cross-Validated RMSE: {rf_rmse}')
print(f'XGBoost Cross-Validated RMSE: {xgb_rmse}')
print(f'Stacked Model Cross-Validated RMSE: {stacked_rmse}')
```

```python
# Make predictions on the test set using the stacked model
stacked_model.fit(X_preprocessed, y)

stacked_predictions = stacked_model.predict(X_test_preprocessed)


```

```python
# Assuming you used log transformation during training
# Inverse log transformation to denormalize
denormalized_predictions = np.exp(stacked_predictions)

# Create a DataFrame with the test set and denormalized predictions
submission_df = pd.DataFrame({
    'Id': test['Id'],
    'SalePrice': denormalized_predictions
})

# Save the DataFrame to a CSV file
submission_df.to_csv('/home/submission.csv', index=False)

# Print a message indicating that the file has been saved
print('Denormalized predictions saved to submission.csv')
```
